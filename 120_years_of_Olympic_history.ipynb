### T-test for Age Difference Between Medal Winners and Non-Winners

```python
# T-test for age difference between medal winners and non-winners
from scipy.stats import ttest_ind
gold_medal_ages = data[data['Medal'] == 'Gold']['Age']
no_medal_ages = data[data['Medal'].isna()]['Age']

t_stat, p_value = ttest_ind(gold_medal_ages, no_medal_ages, nan_policy='omit')
print(f'T-statistic: {t_stat}, P-value: {p_value}')
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(gold_medal_ages, label='Gold Medal Winners', shade=True, color='gold')
sns.kdeplot(no_medal_ages, label='Non-Medalists', shade=True, color='blue')
plt.title('Age Distribution of Gold Medal Winners vs Non-Medalists', fontsize=10)
plt.xlabel('Age', fontsize=10)
plt.ylabel('Density', fontsize=10)
plt.legend()
plt.show()
```

# Section 4: Predictive Modeling

## Predict an athlete's likelihood of winning a medal based on various factors, such as age, weight, and country of origin.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import classification_report
```

```python
X = data[['Age', 'Height', 'Weight']]
y = data['Medal'].notna().astype(int) 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
rf_y_pred = rf_model.predict(X_test)

rf_report = classification_report(y_test, rf_y_pred, output_dict=True)

rf_report_df = pd.DataFrame(rf_report).transpose()

print("Random Forest Classification Report:")
print(rf_report_df)
```

```python
# Measure the accuracy
rf_accuracy = rf_model.score(X_test, y_test)
print(f"Accuracy of the Random Forest model: {rf_accuracy}")
```

```python
# Predict the probability of winning a medal for the new athlete (Age: 28, Height: 194, Weight: 96)
new_athlete = [[28, 194, 96]]
rf_probability = rf_model.predict_proba(new_athlete)

# Output the prediction and probability
print(f"Probability of Winning a Medal: {rf_probability[0][1]:.2f}")
```

```python
new_athlete2 = [[40, 194, 96]]
rf_probability = rf_model.predict_proba(new_athlete2)

# Output the prediction and probability
print(f"Probability of Winning a Medal: {rf_probability[0][1]:.2f}")
```
